```python
import rasterio
import matplotlib.pyplot as plt
import os

with rasterio.open("sldem.tif") as src:
    # Downsample to 1/8 resolution
    scale = 8
    new_height = src.height // scale
    new_width = src.width // scale

    # Read the raster at reduced resolution
    data = src.read(
        1,
        out_shape=(new_height, new_width),
        resampling=rasterio.enums.Resampling.average
    )

    bounds = src.bounds

# Plot
plt.figure(figsize=(12, 6))
plt.imshow(
    data,
    cmap="terrain",
    extent=[bounds.left, bounds.right, bounds.bottom, bounds.top],
    origin="upper"
)
plt.colorbar(label="Elevation (km)")
plt.title("Downsampled SLDEM Elevation Map")
plt.xlabel("Longitude (°E)")
plt.ylabel("Latitude (°N)")
plt.tight_layout()
plt.show()
```

```python
from os import mkdir, path
import os
import requests
from bs4 import BeautifulSoup
import json
from tqdm import tqdm
import pvl
import pandas as pd
from pandas.core.series import Series

BARE_URL="https://pds.lroc.asu.edu/data"
BASE_URL=f"{BARE_URL}/LRO-L-LROC-2-EDR-V1.0/"


def download_file_with_progress(url, output_path):
    if path.exists(output_path):
        print(f"File {output_path} already exists. Skipping download.")
        return
    response = requests.get(url, stream=True)
    response.raise_for_status()

    total_size = int(response.headers.get("content-length", 0))
    block_size = 1024  # 1 Kibibyte

    with (
        open(output_path, "wb") as file,
        tqdm(
            desc=output_path,
            total=total_size,
            unit="iB",
            unit_scale=True,
            unit_divisor=1024,
        ) as bar,
    ):
        for data in response.iter_content(block_size):
            file.write(data)
            bar.update(len(data))

def normalize_lon(lon):
    """Normalize longitude from [0, 360] to [-180, 180]."""
    return lon - 360 if lon > 180 else lon

def get_trek_api_bbox(box1, box2):
    # Normalize longitudes
    lons = [
        normalize_lon(box1["UPPER_LEFT_LONGITUDE"]),
        normalize_lon(box1["LOWER_LEFT_LONGITUDE"]),
        normalize_lon(box1["UPPER_RIGHT_LONGITUDE"]),
        normalize_lon(box1["LOWER_RIGHT_LONGITUDE"]),
        normalize_lon(box2["UPPER_LEFT_LONGITUDE"]),
        normalize_lon(box2["LOWER_LEFT_LONGITUDE"]),
        normalize_lon(box2["UPPER_RIGHT_LONGITUDE"]),
        normalize_lon(box2["LOWER_RIGHT_LONGITUDE"]),
    ]
    lats = [
        box1["UPPER_LEFT_LATITUDE"], box1["UPPER_RIGHT_LATITUDE"],
        box1["LOWER_LEFT_LATITUDE"], box1["LOWER_RIGHT_LATITUDE"],
        box2["UPPER_LEFT_LATITUDE"], box2["UPPER_RIGHT_LATITUDE"],
        box2["LOWER_LEFT_LATITUDE"], box2["LOWER_RIGHT_LATITUDE"],
    ]

    return {
        "ulx": min(lons),   # west
        "uly": max(lats),   # north
        "lrx": max(lons),   # east
        "lry": min(lats),   # south
    }
def get_sldem_image(row,path):
    # Collect all latitudes and longitudes
    lats = [
        row.UPPER_LEFT_LATITUDE,
        row.UPPER_RIGHT_LATITUDE,
        row.LOWER_RIGHT_LATITUDE,
        row.LOWER_LEFT_LATITUDE
    ]
    lons = [
        row.UPPER_LEFT_LONGITUDE,
        row.UPPER_RIGHT_LONGITUDE,
        row.LOWER_RIGHT_LONGITUDE,
        row.LOWER_LEFT_LONGITUDE,
    ]

    ulx = min(lons)  # west
    uly = max(lats)  # north
    lrx = max(lons)  # east
    lry = min(lats)  # south

    base_sldem_url = "https://trek.nasa.gov/moon/TrekWS/rest/transform/latlon/subset/stream/tiff"
    sldem_url = f"{base_sldem_url}?ulx={ulx}&uly={uly}&lrx={lrx}&lry={lry}"
    print(f"Downloading {sldem_url}")
    return
    download_file_with_progress(sldem_url,os.path.join(path,"sldem.tif"))
    print(f"Downloaded {path}/sldem.tif")



def process_dataframe(df):
    def is_nac_image(row):
        name_path = row["FILE_SPECIFICATION_NAME"]
        id=row["PRODUCT_ID"]
        is_nac_map=("NAC" in name_path) and ("MAP" in name_path)
        is_not_cent= "CE" not in id
        return is_nac_map and is_not_cent

    # Columns to retain
    columns_to_keep = [
        'FILE_SPECIFICATION_NAME',
        'PRODUCT_ID',
        'UPPER_LEFT_LATITUDE', 'UPPER_LEFT_LONGITUDE',
        'UPPER_RIGHT_LATITUDE', 'UPPER_RIGHT_LONGITUDE',
        'LOWER_RIGHT_LATITUDE', 'LOWER_RIGHT_LONGITUDE',
        'LOWER_LEFT_LATITUDE', 'LOWER_LEFT_LONGITUDE'
    ]

    # Filter rows first
    filtered_rows = df[df.apply(is_nac_image, axis=1)]

    # Select only desired columns
    filtered_df = filtered_rows[columns_to_keep].copy()

    #for brigtness control
    # filtered_df=(df[
    # (df['INCIDENCE_ANGLE'] < 40) &
    # (df['EMISSION_ANGLE'] < 10) &
    # (df['PHASE_ANGLE'] < 60)]).copy()

    # Deduplicate based on cleaned path
    deduplicated_df = filtered_df.drop_duplicates(subset='FILE_SPECIFICATION_NAME', keep='first')
    return deduplicated_df

def save_coord_json(row1:Series,row2:Series,path:str):
  rows=[row1.to_dict(),row2.to_dict()]
  bbox=get_trek_api_bbox(rows[0],rows[1])
  rows.append(bbox)
  for json_data in (rows):
    name=json_data.get('PRODUCT_ID')
    name=name if name else "bbox"
    with open(f"{os.path.join(path,name)}.json", "w") as json_file:
      json.dump(json_data, json_file)
  return bbox


def get_headers_lbl(path):
  label = pvl.load(path)
  return [i[1]["NAME"] for i in label['INDEX_TABLE'][6:]]

def get_index_tab(index):
  base_path=os.path.join("DATA",index)
  os.makedirs(base_path, exist_ok=True)
  print(f"Created a folder at {base_path}")

  lbl_path=os.path.join(base_path,"INDEX.LBL")
  tab_path=os.path.join(base_path,"INDEX.TAB")
  print(f"{BASE_URL}{index}/INDEX/INDEX.LBL")
  download_file_with_progress(f"{BASE_URL}{index}/INDEX/INDEX.LBL",lbl_path)
  print(f"Downloaded {index}/INDEX/INDEX.LBL")

  print(f"Downloading {index}/INDEX/INDEX.TAB")
  download_file_with_progress(f"{BASE_URL}/{index}/INDEX/INDEX.TAB",tab_path)
  print(f"Downloaded {index}/INDEX/INDEX.TAB")

  headers=get_headers_lbl(lbl_path)
  df = pd.read_csv(tab_path, sep=",", header=None, names=headers)
  return process_dataframe(df)


def get_stereo(url_path,index,id):
    id=id.strip()
    url=f"{BARE_URL}/{url_path.strip()}"

    base_path=os.path.join("DATA",index)
    id_dir_path=os.path.join(base_path,id[:-2])

    os.makedirs(id_dir_path,exist_ok=True)
    id_path=os.path.join(id_dir_path,os.path.basename(url))
    download_file_with_progress(url, f"{id_path}")
    download_file_with_progress(f"{url[:-3]}xml", f"{id_path[:-3]}xml")
    return id_dir_path

def get_dir(name):
  temp=name.strip().split("/")
  id=temp[-1][:-4]
  ext="LE" if id.endswith("RE") else "RE"
  return os.path.join(temp[1],id),id[:-2],ext

def download_images(index, max_samples=5):
    """ index -> LROLRC_0001,LROLRC_0002,LROLRC_0003 ...
        refer -> https://pds.lroc.asu.edu/data/LRO-L-LROC-2-EDR-V1.0/ for more info
    """
    track = set()
    df = get_index_tab(index)

    for idx, (i, head) in enumerate(df.iterrows()):
        head_id = head["PRODUCT_ID"].strip()

        if "NAC" not in head.FILE_SPECIFICATION_NAME or head_id in track:
            continue

        head_url = head["FILE_SPECIFICATION_NAME"]
        path, id, ext = get_dir(head_url)
        tail = df[df['PRODUCT_ID'].str.strip() == f'{id}{ext}']
        if tail.empty:
            continue

        tail_url = tail.iloc[0]["FILE_SPECIFICATION_NAME"]
        tail_id = tail.iloc[0]["PRODUCT_ID"].strip()

        head_dir_path=get_stereo(head_url,index,head_id)
        tail_dir_path=get_stereo(tail_url,index,tail_id)
        bbox=save_coord_json(head,tail.iloc[0],os.path.dirname(head_dir_path))
        print(bbox)
        track.add(tail_id)

        if len(track) >= 2 * max_samples:
            break


download_images("LROLRC_0006",2)
```

```python
# DATA/LROLRC_0001/M107689129/M107689129RE.IMG

from pds4_tools import pds4_read
import matplotlib.pyplot as plt
import numpy as np

# Load the XML labels (which link to their respective IMG files)
base = "DATA/LROLRC_0005/M139237644/M139237644"

img1 = pds4_read(f"{base}LE.xml")[1].data
img2 = pds4_read(f"{base}RE.xml")[1].data

# Check if shapes match
if img1.shape != img2.shape:
    raise ValueError("Images must be the same size to compute difference.")

# Normalize both images
img1_norm = (img1 - img1.min()) / (img1.max() - img1.min())
img2_norm = (img2 - img2.min()) / (img2.max() - img2.min())

# Compute absolute difference
diff = np.abs(img1_norm - img2_norm)

# Plot the difference
plt.figure(figsize=(12, 10))
plt.imshow(diff, cmap='hot')  # You can also try 'seismic' or 'gray'
plt.title("Difference between LE and RE NAC images")
plt.axis('off')
plt.colorbar(label="Difference")
plt.show()


# # Extract image array
# img1 = data[1].data

# # Display it
# plt.figure(figsize=(20, 20))  # Increase display window size
# plt.imshow(img1, cmap='gray', vmin=img1.min(), vmax=img1.max())
# plt.title("LROC NAC Full Image")
# plt.axis('off')
# plt.show()
```
